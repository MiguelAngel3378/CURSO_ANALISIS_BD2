```python
for col in ["Ranked2021", "Ranked2020", 'Country', 'Sector', 'FoundingYear']:
    print(f"\n{col} unique : {data[col].unique()}")
    print(f"{col} value counts : \n{data[col].value_counts()}")
    
    plt.figure(figsize=(20, 7))
    sns.countplot(x = col, data = data, palette = 'hls')
    plt.xticks(rotation=90)
    plt.show()
    
    if col in ["Ranked2021", "Ranked2020"]:
        label = data[col].value_counts()
        
        explode = (0.1, 0.1)
        plt.figure(figsize=(14, 10))
        patches, texts, pcts = plt.pie(label, labels = label.index, colors = ['blue', 'red'],
                                      pctdistance = 0.7,
                                      shadow = True,
                                      startangle = 90,
                                      explode = explode,
                                      autopct = '%1.1f%%',
                                      textprops = {'fontsize':25,
                                                 'color':'black',
                                                 'weight':'bold',
                                                 'family':'serif'})
#         plt.setp(pcts, color = 'white')
        
        hfont = {'fontname':'serif', 'weight':'bold'}
        plt.title(col, size=20, **hfont)
        
        centre_circle = plt.Circle((0,0), 0.40, fc='white')
        
        fig = plt.gcf()
        fig.gca().add_artist(centre_circle)
        plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly import tools
from plotly.subplots import make_subplots
from plotly.offline import iplot, init_notebook_mode
```

```python
plt.figure(figsize=(20,7))
plt.title('2020 vs 2021',fontsize = 20)
sns.lineplot(data=data['Ranked2020'],linewidth = 2, label = '2020')
sns.lineplot(data=data['Ranked2021'],linewidth = 2, label = '2021', color='green')
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.title('Revenue 2020 vs Revenue 2017',fontsize = 20)
sns.lineplot(data=data['Revenue2020'],linewidth = 2, label = '2020')
sns.lineplot(data=data['Revenue2017'],linewidth = 2, label = '2017', color='r')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=data, x='FoundingYear', y='CAGR', color='black', alpha=0.9)
plt.title('CAGR vs FoundingYear',fontsize = 20)
plt.show()
```

```python
sns.jointplot(x='CAGR', y='FoundingYear', data=data , height = 10 , kind='hex')
plt.title('CAGR vs FoundingYear',fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.scatterplot(x='Sector',y='Country',hue='CAGR',data=data)
plt.xticks(rotation=90)
plt.title('CAGR vs ountry & Sector',fontsize = 20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,12))
plt.xticks(rotation=90)
ax = sns.stripplot(y='Country', x='Sector', data=data, palette='Pastel1', s=10, marker='P', linewidth=1)
plt.title('CAGR vs ountry & Sector',fontsize = 20)
plt.show()
```

```python
fig= px.histogram(data, x='Sector', color='Sector',y='Revenue2020',
                  marginal='box',hover_data=data.columns,
                  title= 'Sector revenues for 2020', width=1000, height=500)
fig.show()
```

```python
plt.figure(figsize=(18,12))
sns.scatterplot(data=data, x='Sector', y='Revenue2020', hue='Country', legend=True)
plt.xticks(rotation=90)
plt.title('Revenue for 2020 vs countries & sectors',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(18,12))
sns.scatterplot(data=data, x='Sector', y='Revenue2017', hue='Country', legend=True)
plt.xticks(rotation=90)
plt.title('Revenue for 2017 vs countries & sectors',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(x=data['CAGR'],y=data['Revenue2020'],fill=True,cmap='Reds', thresh=0.05)
plt.title('CAGR vs Revenue2020',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(x=data['CAGR'],y=data['Revenue2017'],fill=True,cmap='Reds', thresh=0.05)
plt.title('CAGR vs Revenue2017',fontsize = 20)
plt.show()
```

```python
selected_years = [2016, 2014, 2001, 2007, 1999, 1984, 1952, 1975, 1992]
df_selected = data[data['FoundingYear'].isin(selected_years)]

fig = px.choropleth(
    data_frame=df_selected,
    locations='Country',
    locationmode='country names',
    color='Revenue2020',
    color_continuous_scale='Plasma',
    range_color=(df_selected['Revenue2020'].min(), df_selected['Revenue2020'].max()),
)

fig.update_layout(
    title='Revenue distribution relative to countries for 2020 on the map',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular',
        bgcolor='black',
        landcolor='white')
)

fig.show()
```
