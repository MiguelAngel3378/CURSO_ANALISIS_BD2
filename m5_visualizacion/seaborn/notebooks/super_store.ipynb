```python
numeric_columns = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

for column in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=column)
    plt.title(f'Hisplot de {column}')
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    plt.xlim(lower_limit, upper_limit)
    
    plt.show()
```

```python
numeric_columns = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

for column in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=df, x=column)
    plt.title(f'Hisplot de {column}')
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    plt.xlim(lower_limit, upper_limit)
    
    plt.show()
```

BARPLOT : VENTAS TOTALES POR AO

```python
# 1  CREAR COLUMNA YEAR
# 2  GROUP BY DE SALES POR YEAR / GROUPBY DE PROFIT POR YEAR
# 3  BARPLOT
```

```python
# 1  CREAR COLUMNA YEAR
df = pd.read_csv('../../../data/varios/superstore_dataset2011-2015.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['year'] = df['Order Date'].dt.year
```

```python
# 2  GROUP BY DE SALES POR YEAR / GROUPBY DE PROFIT POR YEAR
sales_per_year = df.groupby('year')['Sales'].sum()
profit_per_year = df.groupby('year')['Profit'].sum()
```

```python
# VISUALISAR DATOS 
sales_per_year.values
```

```python
# 3  BARPLOT
sns.barplot(x=sales_per_year.index, y=sales_per_year.values)
```

```python
sns.barplot(x=profit_per_year.index, y=profit_per_year.values)
```

```python
# CREAR DATAFRAME DE SALES POR YEAR Y MOSTRAR GRAFICA DE AREA
df_sales_per_year = pd.DataFrame({
    'Sales': sales_per_year.values,    #SERIE
    'Profit': profit_per_year.values   #SERIE
}, index=sales_per_year.index)
    
df_sales_per_year.plot(kind='area', alpha=0.5)
plt.title('Ventas y Beneficios por ao')
```

```python
df_sales_per_year.plot(kind='bar', alpha=0.5)
plt.title('Ventas y Beneficios por ao')
```

```python
# SEABORN
fig, ax1 = plt.subplots(figsize=(10,6))

# BARPLOT PARA SALES PER YEAR
sns.barplot(x=sales_per_year.index, y=sales_per_year.values, ax=ax1, alpha=0.5)
ax1.set_ylabel('Sales')
plt.ticklabel_format(style='plain', axis='y')

# BARPLOT PARA PROFIT PER YEAR
ax2= ax1.twinx()   # EJE X COMPARTIDO
sns.barplot(x=profit_per_year.index, y=profit_per_year.values, ax=ax2, alpha=0.5)
ax2.set_ylabel('Profit')
```

```python
fig, ax1 = plt.subplots()

color = 'blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Sales', color=color)
ax1.bar(df_sales_per_year.index, df_sales_per_year['Sales'], color=color) # BARRAS
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'red'
ax2.set_ylabel('Profit', color=color)
ax2.plot(df_sales_per_year.index, df_sales_per_year['Profit'], color=color) # LINEAS
ax2.tick_params(axis='y', labelcolor=color)
```
